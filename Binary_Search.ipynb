```python
import time
import random
import matplotlib.pyplot as plt

# Recursive Binary Search Function
def binary_search_recursive(arr, target, low, high):
    if low > high:
        return -1
    mid = (low + high) // 2
    if arr[mid] == target:
        return mid
    elif arr[mid] > target:
        return binary_search_recursive(arr, target, low, mid - 1)
    else:
        return binary_search_recursive(arr, target, mid + 1, high)

# Function to measure search time
def measure_time(n):
    arr = sorted(random.sample(range(1, n * 10), n))  # unique sorted array
    target = random.choice(arr)  # pick a random element to search
    start_time = time.perf_counter()
    binary_search_recursive(arr, target, 0, len(arr) - 1)
    end_time = time.perf_counter()
    return end_time - start_time

# Main experiment
sizes = [10, 100, 500, 1000, 5000, 10000, 50000, 100000]
times = []

for n in sizes:
    elapsed_time = measure_time(n)
    times.append(elapsed_time)
    print(f"n={n}, time={elapsed_time:.8f} seconds")

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(sizes, times, marker='o', linestyle='-', color='b')
plt.title('Recursive Binary Search: Time vs Input Size')
plt.xlabel('Number of Elements (n)')
plt.ylabel('Time Taken (seconds)')
plt.grid(True)
plt.show()
```
